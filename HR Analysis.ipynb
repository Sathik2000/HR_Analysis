```python
sathik
```

# Find Total Working Days

```python
sathik['Hire Date']=pd.to_datetime(sathik['Hire Date'])
```

```python
sathik['Exit Date']=pd.to_datetime(sathik['Exit Date'])
```

```python
sathik['working_days'] = sathik['Exit Date'] - sathik['Hire Date']
```

```python
sathik
```

# Rename Annual Salary column

```python
sathik.rename(columns={'Annual Salary':'Annual Salary($)'})
```

# which department worker worked more days

```python
D_W_days = pd.DataFrame({'Department': sathik['Department'], 'working_days': sathik['working_days']})
```

```python
D_W_days
```

```python
D_W_days = D_W_days.sort_values('working_days', ascending=False).reset_index(drop=True)
```

```python
D_W_days
```

```python
 group=D_W_days.groupby('Department')['working_days'].sum()
```

```python
group
```

```python
group_data = pd.DataFrame(group).reset_index()
```

```python
group_data
```

```python
group_data=group_data.sort_values('working_days', ascending=False).reset_index(drop=True)
```

```python
group_data
```

```python
group_data.dtypes
```

```python
group_data['working_days'] = group_data['working_days'].dt.days.astype(int)
```

```python
group_data.dtypes
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
plt.bar(group_data['Department'],group_data['working_days'],color='darkblue')
plt.xlabel('Departments')
plt.ylabel('Working Days')
plt.title('Depart Past Employee with working days')
```
